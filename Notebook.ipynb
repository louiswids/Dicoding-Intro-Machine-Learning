```python
# Import Libarary
import numpy as np
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
import zipfile,os
import random
from tensorflow.keras.preprocessing.image import ImageDataGenerator

print(tf.__version__)
```

```python
# Akuisisi dataset
!wget --no-check-certificate \
  https://github.com/dicodingacademy/assets/releases/download/release/rockpaperscissors.zip \
  -O /tmp/rockpaperscissors.zip
```

```python
# Unzipping dataset
local_zip = '/tmp/rockpaperscissors.zip'
zip_ref = zipfile.ZipFile(local_zip, 'r')
zip_ref.extractall('/tmp')
zip_ref.close()
 
# Setting base directory
base_dir = '/tmp/rockpaperscissors/rps-cv-images'
```

```python
# Verifikasi konten dari zip
os.listdir(base_dir)
```

```python
# Data augmentation dan splitting
train_test_datagen = ImageDataGenerator(
                    rescale=1./255,
                    rotation_range=20,
                    horizontal_flip=True,
                    shear_range = 0.2,
                    fill_mode = 'nearest',
                    validation_split = 0.4)
```

```python
# Creating training data and testing data
train_generator = train_test_datagen.flow_from_directory(
        base_dir,
        subset='training',
        target_size=(150, 250),# Ukuran gambar ditetapkan sebagai 150x250 pixel
        batch_size=32, # Menetapkan ukuran batch gambar sebagai 32
        color_mode='grayscale',# Channel warna ditetapkan sebagai grayscale untuk mengurangi komputasi
        class_mode='categorical' # Menentukan metode inferensi sebagai kategorikal (x,x,x)
        )
 
test_generator = train_test_datagen.flow_from_directory(
        base_dir,
        subset='validation',
        target_size=(150, 250),# Ukuran gambar ditetapkan sebagai 150x250 pixel
        batch_size=32, # Menetapkan ukuran batch gambar sebagai 32
        color_mode='grayscale',# Channel warna ditetapkan sebagai grayscale untuk mengurangi komputasi
        class_mode='categorical'# Menentukan metode inferensi sebagai kategorikal (x,x,x)
        )
```

```python
# Menunjukkan beberapa sampel data dari setiap batch
plt.subplot(221)
plt.imshow(train_generator[0][0][0])
plt.subplot(222)
plt.imshow(train_generator[0][0][2])
plt.subplot(223)
plt.imshow(train_generator[0][0][5])
plt.subplot(224)
plt.imshow(train_generator[0][0][10])

plt.show()
```

```python
# Penyusunan arsitektur model
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 250, 1)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.3),    
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),   
    tf.keras.layers.Conv2D(216, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),    
    tf.keras.layers.Flatten(), # layer flatten untuk mengubah data agar dapat diproses dense layer
    tf.keras.layers.Dense(216, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(64, activation='relu'),    
    tf.keras.layers.Dense(3, activation='softmax')
])
```

```python
# Compiling model dan ringkasan model
model.compile(loss='categorical_crossentropy',
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])
model.summary()
```

```python
# Proses fitting model
# Validation dan training steps tidak dispesifikasi agar semua gambar dapat dimasukkan pada proses fitting

# Penggunaan early stopping membuat model tidak optimal
# callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience = 4)
history = model.fit(
        train_generator,
        epochs=30, # tambahkan epochs jika akurasi model belum optimal
        validation_data=test_generator, # menampilkan akurasi pengujian data validasi
        # callbacks = [callback]
        verbose=2)
```

```python
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

train_loss = history.history['loss']
val_loss = history.history['val_loss']

fig,ax=plt.subplots(2, figsize=(10,10))
ax[0].plot(train_accuracy)
ax[0].plot(val_accuracy)
ax[0].set_title('Akurasi pada Setiap Epoch')
ax[0].set_xlabel('Epoch')
ax[0].set_ylabel('Akurasi')
ax[0].legend(['training', 'valdation'])

ax[1].plot(train_loss)
ax[1].plot(val_loss)
ax[1].set_title('Loss pada Setiap Epoch ')
ax[1].set_xlabel('Epoch')
ax[1].set_ylabel('Loss')
ax[1].legend(['training', 'validation'])
```

```python
# Testing pada dataset baru
hand = cv2.imread('hand.jpeg') #Data dapat diganti menjadi data lain yang akan digunakan untuk testing
hand_resized = cv2.resize(hand, (250,150))
hand_resized=cv2.cvtColor(hand_resized, cv2.COLOR_BGR2GRAY)
plt.imshow(hand_resized)
x = np.array(hand_resized)
x = np.expand_dims(x, axis=0)
```

```python
# Membuat prediksi dan mengetes hasilnya
predict = model.predict(x)
  
if (predict[0][0] == 1) :
    print("Kertas")
elif (predict[0][1] == 1) :
    print("Batu")
elif (predict[0][2] == 1) :
    print("Gunting")
```

## Kesimpulan dan Analysis Proses Training
Ada beberapa kesimpulan dan analisis dari proses training model klasifikasi citra batu gunting kertas:
